```python
# Import packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset into dataframe

data = pd.read_excel("C:/Users/Azhar/Desktop/Python/Election Poll 2024/election poll_01.xlsx")
```

```python

# Calculate the percentage of each political party
party_percentage = data['Political_Party'].value_counts(normalize=True) * 100

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=party_percentage.index, y=party_percentage.values, palette='viridis')

# Display percentage on data labels
for index, value in enumerate(party_percentage):
    plt.text(index, value, f'{value:.2f}%', ha='center', va='bottom')

# Customize the chart
plt.title("Q.1 In your opinion, which political party is likely to receive significant support in the upcoming Lok Sabha election?")
plt.xlabel("Political Party")
plt.ylabel("Percentage Support")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Calculate the percentage of each political coalition
coalition_percentage = data['Coalition'].value_counts(normalize=True) * 100

# Create a donut chart
plt.figure(figsize=(5, 5))
plt.pie(coalition_percentage, labels=coalition_percentage.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'), wedgeprops=dict(width=0.4))

# Draw a circle at the center to create a donut chart
centre_circle = plt.Circle((0, 0), 0.6, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Customize the chart
plt.title("Q.2 Which political coalition do you think will garner considerable support in the upcoming election?")
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle

# Show the plot
plt.show()
```

```python

# Calculate the percentage of each choice for Prime Minister
pm_choice_percentage = data['PM_Choice'].value_counts(normalize=True) * 100

# Create a bar chart using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=pm_choice_percentage.index, y=pm_choice_percentage.values, palette='Set2')

# Display percentage on data labels
for index, value in enumerate(pm_choice_percentage):
    plt.text(index, value, f'{value:.2f}%', ha='center', va='bottom')

# Customize the chart
plt.title("Q.3 In your view, who do you believe is a preferred choice for the Prime Minister in the 2024 election?")
plt.xlabel("Preferred Prime Minister")
plt.ylabel("Percentage Support")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Calculate the percentage of each response regarding the importance of the party face
party_face_percentage = data['Party_Face'].value_counts(normalize=True) * 100

# Create a pie chart
plt.figure(figsize=(5, 5))
plt.pie(party_face_percentage, labels=party_face_percentage.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))

# Customize the chart
plt.title("Q.4 Do you believe that the face of the party (the Prime Ministerial candidate) is important for the upcoming Lok Sabha election?")
plt.show()
```

```python
# Load dataset into dataframe

data02 = pd.read_excel("C:/Users/Azhar/Desktop/Python/Election Poll 2024/election poll_02.xlsx")
```

```python
data02.head()
```

```python

# Assuming data02 is your DataFrame
# If you haven't loaded the data yet, you can load it using pd.read_excel('your_excel_file.xlsx')
# For example: data02 = pd.read_excel('election_poll_02.xlsx')

# Specify the columns to keep as is and columns to unpivot
id_vars = ['Sr.No.']
value_vars = data02.columns[1:]  # Assuming Option 1 to Option 7 are the columns to unpivot

# Unpivot the DataFrame
unpivoted_data02 = pd.melt(data02, id_vars=id_vars, value_vars=value_vars, var_name='Option', value_name='Response')

# Display the unpivoted DataFrame
print(unpivoted_data02)
```

```python
# Remove rows with NaN values in the 'Response' column
data02_Cleaned = unpivoted_data02.dropna(subset=['Response'])
```

```python
data02_Cleaned .head(25)
```

```python
# Save the cleaned DataFrame to an Excel file
data02_Cleaned .to_excel('data02_Cleaned .xlsx', index=False)
```

```python
# Count distinct Sr.No. values
distinct_sr_no_count = data02_Cleaned ['Sr.No.'].nunique()

# Display the count
print(f"Count of Distinct Sr.No.: {distinct_sr_no_count}")
```

```python

# Calculate the total count for each response
total_count = data02_Cleaned .groupby('Response').size().reset_index(name='TotalCount')

# Display the total count
print(total_count)
```

```python

# Calculate the total count for each response
total_count = data02_Cleaned .groupby('Response').size().reset_index(name='TotalResponse')

# Calculate the overall total count
overall_total = total_count['TotalResponse'].sum()

# Display the overall total count
print(f"Overall Total Response Count: {overall_total}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a countplot
plt.figure(figsize=(12, 6))

# Sort the values in descending order
sorted_order = data02_Cleaned['Response'].value_counts().index
sns.countplot(x='Response', data=unpivoted_data02, order=sorted_order, palette='viridis')

# Display count on data labels
for index, value in enumerate(data02_Cleaned ['Response'].value_counts()[sorted_order]):
    plt.text(index, value, f'{value}', ha='center', va='bottom')

# Customize the chart
plt.title("Distribution of Responses count for each Option (Important Issues)")
plt.xlabel("Important Issues")
plt.ylabel("Count of Each Response")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Calculate the percentage of total count of responses divided by count of distinct Sr.No.
percentage_gt_count = (data02_Cleaned .groupby('Response').size() / distinct_sr_no_count) * 100

# Sort the values in descending order
percentage_gt_count = percentage_gt_count.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=percentage_gt_count.index, y=percentage_gt_count.values, palette='viridis')

# Display percentage on data labels
for index, value in enumerate(percentage_gt_count):
    plt.text(index, value, f'{value:.2f}%', ha='center', va='bottom')

# Customize the chart
plt.title("Q.8 In your perception, which of the following issues do you consider essential or important when the public decides their vote in the upcoming Lok Sabha election?")
plt.xlabel("Important Issues")
plt.ylabel("Percentage Distribution")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
